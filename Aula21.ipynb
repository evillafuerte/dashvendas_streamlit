### A biblioteca glob é uma biblioteca padrão do Python que fornece uma maneira simples de encontrar arquivos que correspondam a padrões específicos em um determinado diretório. Ela permite que você liste os arquivos que atendem a um padrão específico, facilitando a busca e a manipulação desses arquivos.

```python
arquivos = sorted(glob("Planilhas/*.xlsx"))
```

```python
arquivos
```

```python
todos_arquivos = pd.concat((pd.read_excel(arquivo) for arquivo in arquivos))
```

```python
todos_arquivos
```

```python
todos_arquivos.dtypes
```

```python
todos_arquivos["Receita"] = todos_arquivos["Vendas"] * todos_arquivos["Qtde"]
```

```python
todos_arquivos
```

```python
todos_arquivos["Ano_Vend"] = todos_arquivos["Data"].dt.year
```

```python
todos_arquivos
```

```python
todos_arquivos.to_excel("planilha_vendas.xlsx", index=False)
```

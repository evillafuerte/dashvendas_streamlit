## Melt

* id_vars: Variáveis que serão mantidas como identificadores únicos (colunas que não serão derretidas).
* value_vars: Colunas a serem derretidas em linhas. Se não for especificado, todas as colunas restantes serão usadas.
* var_name: Nome da coluna que conterá os nomes das variáveis derretidas.
* value_name: Nome da coluna que conterá os valores correspondentes das variáveis derretidas.

```python
notas = df.melt(id_vars="Nome", var_name="Disciplina", value_name="Nota")
```

```python
notas
```

```python
notas.groupby(["Nome", "Disciplina"])["Nota"].sum()
```
